## Fig.6(b)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10000
L = 50
Tinit = 1000 # washout time step
Tobs = 10000 # simulation time step
max_delay = 100 # max delay time for calculating MC
g = 0.6
sigma = 0.1
eta_list = [-0.65, 0.65, -0.5, 0.5, 0]
phi = np.tanh
nums = 10
```

```python
data = np.load("data/memorycurve_reciprocal.npy") # shape: [len(eta_list), nums, max_delay]
```

```python
fig = plt.figure(figsize=(7,5))
markers =["o", "x", "^", "D", "s"]
colors = ["red", "orange", "blue", "cyan", "k"]
i=0
for i in range(len(eta_list)):
    mean = np.mean(data[i], axis=0)
    std = np.std(data[i], axis=0)
    plt.plot(np.arange(1, max_delay+1), mean, markersize=6, marker=markers[i], label=r"$\eta=$"+f"{eta_list[i]}", color=colors[i])
    plt.fill_between(np.arange(1, max_delay+1), mean-std, mean+std, alpha=0.5, color=colors[i])
plt.xlabel(r"$d$",fontsize=25)
plt.ylabel(r"$M_{d}$",fontsize=25)
plt.tick_params(labelsize=13)
plt.ylim(0,1.1)
plt.xlim(-5, 80)
plt.legend(fontsize=20)
plt.grid(linewidth=0.2)
plt.show()
```
