```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10000
Tinit = 1000 # washout time step
Tobs = 10000 # simulation time step
max_delay = 100 # max delay time for calculating MC
g = 0.6
sigma = 0.1
eta_list = [-0.65, 0.65, -0.5, 0.5, 0]
phi = np.tanh
nums = 10
```

```python
autocoef = np.load("data/autocoef_reciprocal.npy") #shape: [lem(eta_list), nums, max_delay+1]
```

```python
fig = plt.figure(figsize=(7,5))
markers =["o", "x", "^", "D", "s"]
colors = ["red", "orange", "blue", "cyan", "k"]

for ideta, eta in enumerate(eta_list):
    mean = np.mean(autocoef[ideta], axis=0)
    std = np.std(autocoef[ideta], axis=0)
    
    plt.plot(np.arange(0, max_delay+1), mean, color=colors[ideta], marker=markers[ideta], label=r"$\eta=$"+f"{eta}")
    plt.fill_between(np.arange(0, max_delay+1), mean-std, mean+std, alpha=0.5)

plt.xticks(np.arange(0,21,5), size=15)
plt.yticks(size=15)
plt.xlim(-1,21)
plt.xlabel(r"$d$", size=25)
plt.ylabel(r"$A(d)$", size=25)
plt.legend(fontsize=15)
plt.grid(linewidth = 0.3)
#plt.title(f"Correlated: N={N}, g={g}, sigma_s={sigma}, sigma_n={0}")
plt.show()
```
